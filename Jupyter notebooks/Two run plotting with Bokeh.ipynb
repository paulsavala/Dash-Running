```python
import config as cfg # My own config.py file with configuration parameters
from models import *

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np

from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
output_notebook()
```

```python
run1 = Run('San_Diego_run_with_Jeff_2')
run2 = Run('Final_San_Diego_run')
```

```python
minutes = cfg.minutes
time = np.arange(1, 60*5+1)
```

```python
def speed_at_gradient(run, g=0):
    return run.speed_at_gradient(g)[1:]
```

```python
p = figure(title="Speed at Gradient", plot_height=300, plot_width=600, y_range=(2,5))
r = p.multi_line([time, time], [speed_at_gradient(run1, 0), speed_at_gradient(run2, 0)],\
                color=['#a92121','#2222aa'], line_width=[3,1])
```

```python
def update_time(seconds):
    updated_time = np.arange(1, seconds)
    r.data_source.data['xs'] = [updated_time, updated_time]
    push_notebook()
    
def update_gradient(gradient):
    r.data_source.data['ys'] = [speed_at_gradient(run1, gradient), speed_at_gradient(run2, gradient)]
    push_notebook()
```

```python
interact(update_time, seconds=widgets.IntSlider(min=1,max=60*10,step=1,value=60*5))
```

```python
interact(update_gradient, gradient=widgets.IntSlider(min=-5,max=5, step=1,value=0))
```

```python
show(p, notebook_handle=True)
```
